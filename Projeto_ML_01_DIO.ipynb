## Exemplo de uso do MATPLOT LIB

- Função: Importa a biblioteca matplotlib que é amplamente utilizada em Python para criar gráficos e visualizações de dados.
- Uso: A importação da biblioteca é necessária para que possamos utilizar suas funções e métodos para criar gráficos e visualizações.

```python
import matplotlib
```

- from sklearn.datasets import make_regression: Importa a função make_regression do módulo datasets da biblioteca scikit-learn. Essa função é usada para gerar um conjunto de dados de regressão sintético, útil para testar algoritmos de regressão.
- x, y = make_regression(n_samples=200, n_features=1, noise=30): Gera um conjunto de dados de regressão com 200 amostras e 1 característica (feature). O parâmetro noise=30 adiciona ruído aleatório aos dados para simular a variabilidade encontrada em dados reais.
  - x: Array contendo os valores das características (features) do conjunto de dados.
  - y: Array contendo os valores alvo (target) associados às características.
- import matplotlib.pyplot as plt: Importa o módulo pyplot da biblioteca matplotlib comumente usado para criar gráficos de maneira simplificada.

```python
from sklearn.datasets import make_regression
x, y = make_regression(n_samples=200, n_features=1, noise=30)
import matplotlib.pyplot as plt
```

- plt.scatter(x, y): Cria um gráfico de dispersão (scatter plot) utilizando os dados gerados. Os valores de x são representados no eixo x e os valores de y são representados no eixo y. Este tipo de gráfico é útil para visualizar a relação entre duas variáveis.
- plt.show(): Exibe o gráfico criado. Esta função é necessária para renderizar e mostrar o gráfico na tela.

```python
plt.scatter(x,y)
plt.show()
```

### Resumo
- A primeira caixa importa a biblioteca matplotlib.
- A segunda caixa gera um conjunto de dados de regressão sintético e importa o módulo pyplot para visualização.
- A terceira caixa cria e exibe um gráfico de dispersão dos dados gerados.

Este conjunto de código é útil para testar e visualizar a relação entre variáveis em um contexto de regressão, onde se tenta ajustar um modelo para prever um valor contínuo com base em uma ou mais características.
